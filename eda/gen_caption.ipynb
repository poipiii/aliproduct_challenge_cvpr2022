```python
import pickle as pkl
import pandas as pd
import os
```

```python
df = pd.read_csv("/home/ubuntu/Desktop/CVPR 2022 AliProducts Challenge/code/data/val_data_gen_cap.csv")
df.head()
```

```python
all_pred = []
caption_folder = "/home/ubuntu/Desktop/CVPR 2022 AliProducts Challenge/code/model/BLIP/output/caption_finetune/results"
for file_num in range(len(os.listdir(caption_folder))):
    with open(f"{caption_folder}/blip_gen_caption_{file_num}.pkl","rb") as file:
        print(file_num)
        captions = pkl.load(file)
        all_pred.append(captions)
captions = [c for l in all_pred for p in l for c in p ]
```

```python
with open("/home/ubuntu/Desktop/CVPR 2022 AliProducts Challenge/gen_captions_val_base.pkl","rb") as file:
    captions_base = pkl.load(file)
captions_base = [c for p in captions_base for c in p]
print(len(captions_base))
```

```python
df["gen_caption_ft"] = captions
df["gen_caption_base"] = captions_base
```
